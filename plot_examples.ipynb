```pyspark
# This code runs in Spark Driver
import pandas as pd
import matplotlib.pyplot as plt
plt.switch_backend('agg')
import numpy as np

labels = 'apples', 'oranges', 'lemons', 'limes'
sizes = [10, 15, 5, 20]

fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct='%1.1f%%')
ax.axis('equal')
```
